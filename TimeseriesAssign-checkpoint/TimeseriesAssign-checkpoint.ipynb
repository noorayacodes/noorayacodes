```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset and parse the date column
df = pd.read_csv(r'C:\Users\hp\AppData\Local\Temp\12dc5ec2-5169-45d2-b312-00597832a905_Timeseries.zip.905\Timeseries\exchange_rate.csv', parse_dates=['date'])

# Set the 'Date' column as the index
df.set_index('date', inplace=True)

# Display basic information
print("Dataset Info:")
print(df.info())
print("\nFirst few rows:")
print(df.head())

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.iloc[:, 0], label='USD to AUD Exchange Rate', color='teal')
plt.title('USD to AUD Exchange Rate Over Time')
plt.xlabel('date')
plt.ylabel('Exchange Rate')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```
